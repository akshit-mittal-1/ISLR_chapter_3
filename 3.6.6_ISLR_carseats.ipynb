# IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import statsmodels.api as sm
import sklearn
from sklearn.linear_model import LinearRegression
```

# LOADING DATA

```python
data = pd.read_csv('Carseats.csv')
```

```python
data
```

# DATA PRE-PROCESSING 

```python
us = pd.get_dummies(data['US'], prefix = 'US')
urban = pd.get_dummies(data['Urban'], prefix = 'Urban')
shelve = pd.get_dummies(data['ShelveLoc'], prefix = 'Shelve')
```

```python
shelve
```

```python
data = data.drop(['US', 'Urban', 'ShelveLoc'], axis=1)
```

```python
data = pd.concat([data, us['US_Yes'], urban['Urban_Yes'], shelve['Shelve_Good'], shelve['Shelve_Medium'] ], axis=1)
```

```python
data
```

```python
data['income*ad'] = data['Income']*data['Advertising']
data['price*age'] = data['Price']*data['Age']
```

```python
data
```

```python
target = data['Sales']
data1 = data.drop(['Sales'], axis=1)
```

# USING SKLEARN

```python
lr = LinearRegression()
lr.fit(data1, target)
```

```python
lr.coef_          #coefficient of weights
```

```python
lr.intercept_     #BIAS TERM
```

# USING STATSMODELS

```python
X = data1      
y = target
X = sm.add_constant(X)     #adding constant 1 to training data

model = sm.OLS(y,X).fit()
```

```python
print(model.summary())
```
