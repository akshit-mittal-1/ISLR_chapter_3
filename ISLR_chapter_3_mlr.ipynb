# MULTIPLE LINEAR REGRESSION USING SKLEARN

```python
target = DataFrame(data['medv'])
data1 = data.drop(['medv'], axis=1)   #DATA WITHOUT target variable
```

```python
data1
```

```python
lr = LinearRegression()
lr.fit(data1, target)
```

```python
lr.coef_    #weights of all the parameters
```

```python
lr.intercept_   #bias term 
```

```python
lr.score(data1, target)   #R^2 score of the model
```

# MULTIPLE LINEAR REGRESSION USING STATSMODELS 

```python
#we use statsmodels as it gives a lot more information i.e. adjusted R squared, F statistic, p values, standard errors of weights and 95% confidence interval 
X = data1      
y = target
X = sm.add_constant(X)     #adding constant 1 to training data

model = sm.OLS(y,X).fit()
```

```python
print(model.summary())
```
